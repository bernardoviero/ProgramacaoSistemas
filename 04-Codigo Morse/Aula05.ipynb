```
textoParaTraduzir = input('Digite o texto para converter para Código Morse:').upper()#lower()
textoSaida = []
tamanho = len(textoParaTraduzir)
indice = 0
while indice < tamanho:
  if textoParaTraduzir[indice]=='A':
    textoSaida.append('.-')
  if textoParaTraduzir[indice]=='B':
    textoSaida.append('-...')
  if textoParaTraduzir[indice]=='C':
    textoSaida.append('-.-.')
  if textoParaTraduzir[indice]=='D':
    textoSaida.append('-..')
  if textoParaTraduzir[indice]=='E':
    textoSaida.append('.')
  if textoParaTraduzir[indice]=='F':
    textoSaida.append('..-.')
  if textoParaTraduzir[indice]=='G':
    textoSaida.append('--.')
  if textoParaTraduzir[indice]=='H':
    textoSaida.append('....')
  if textoParaTraduzir[indice]=='I':
    textoSaida.append('..')
  if textoParaTraduzir[indice]=='J':
    textoSaida.append('.---')
  if textoParaTraduzir[indice]=='K':
    textoSaida.append('-.-')
  if textoParaTraduzir[indice]=='L':
    textoSaida.append('.-..')
  if textoParaTraduzir[indice]=='M':
    textoSaida.append('--')
  if textoParaTraduzir[indice]=='N':
    textoSaida.append('-.')
  if textoParaTraduzir[indice]=='O':
    textoSaida.append('---')
  if textoParaTraduzir[indice]=='P':
    textoSaida.append('.--.')
  if textoParaTraduzir[indice]=='Q':
    textoSaida.append('--.-')
  if textoParaTraduzir[indice]=='R':
    textoSaida.append('.-.')  
  if textoParaTraduzir[indice]=='S':
    textoSaida.append('...')
  if textoParaTraduzir[indice]=='T':
    textoSaida.append('-')
  if textoParaTraduzir[indice]=='U':
    textoSaida.append('..-')
  if textoParaTraduzir[indice]=='W':
    textoSaida.append('.--')
  if textoParaTraduzir[indice]=='X':
    textoSaida.append('-..-')
  if textoParaTraduzir[indice]=='Y':
    textoSaida.append('-.--')
  if textoParaTraduzir[indice]=='Z':
    textoSaida.append('--..')
  if textoParaTraduzir[indice]==' ':
    textoSaida.append('[silencio]')
  indice = indice + 1
print(textoSaida)
```
