CONVERSÃO DECIMAL PARA BINÁRIO

```

numeroBinario= [] #Vetor representando o numero binário vazio.
numeroDecimal = input("Digiteu número: ")
numeroDecimal = int(numeroDecimal)
numero = numeroDecimal
dividido = int(numero / 2)

while dividido != 0:
  dif = numero - (dividido*2)
  numeroBinario.append(dif) #Acrescentando na ultima posição o número binário a diferença. (Append)
  numero = dividido
  #print(numero)
  dividido = int(numero / 2)
numeroBinario.append(dividido+1)
numeroBinario = numeroBinario[::-1]

print(f"O número {numeroDecimal}, convertido para binário é: {numeroBinario}")
```
