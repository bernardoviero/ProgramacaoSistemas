Binária para decimal:


```
# Conversão de base: Binária para decimal:
''' 
101001 -> Lendo da direita para a esquerda

(1x2**5) + (0x2**4) + (1x2**3) + (0x2**2) + (0x2**1) + (1x2**0)

'''
resultado = ((1*2**5) + (0*2**4) + (1*2**3) + (0*2**2) + (0*2**1) + (1*2**0))
print(resultado)
```

Programa de conversão binária.


```
import math

numeroBinario= '101001'
tamanhoBinario = len(numeroBinario) #Mostra o tamanho dos caracteres;
numeroDecimal = 0 #Inicializando em 0

'''
# TamanhoBinario - 1, decrementação, pois:

numeroBinario[0] = 1
numeroBinario[1] = 0
numeroBinario[2] = 1
numeroBinario[3] = 0
numeroBinario[4] = 0
numeroBinario[5] = 1
'''
controle = tamanhoBinario-1

while tamanhoBinario > 0:
  expoente = controle - (tamanhoBinario-1)
  # print(numeroBinario[tamanhoBinario-1],'x2^',expoente)
  numeroCalculado = int(numeroBinario[tamanhoBinario-1]) * math.pow(2,expoente)
  numeroDecimal = numeroDecimal + numeroCalculado
  tamanhoBinario = tamanhoBinario - 1

  print(f'O número binário {numeroBinario}, convertido para decimal é {numeroDecimal}')

```
