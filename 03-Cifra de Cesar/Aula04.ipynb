
**Cifra de Cesar** (Texto normal -> Cifrado)

```
caractere = input('Digite um uma palavra/frase para convertemos a cifra de César: ')

saida = [] # vetor onde será escrita a saída da cifragem

tamanhoCaracter = len(caractere) # len retorna o tamanho de entrada
print(tamanhoCaracter)

indice = 0

#com o laço while, conseguimos percorrer o vetor de caracteres de entrada
while indice < tamanhoCaracter:

  if caractere[indice] == ' ':
    saida.append(' ')

  else:
    valorDoCaracterer = ord(caractere[indice])

    if valorDoCaracterer >= 88:
      convertido = (valorDoCaracterer-26) + 3

    else:
      convertido = valorDoCaracterer + 3
    saida.append(chr(convertido)) #coloca na ultima posição do vetor saida o caracter convertido

  indice = indice + 1

print(saida)
```
