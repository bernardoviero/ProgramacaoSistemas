ENTENDENDO O SPLIT


```
texto = '233.562.489.48,5'

centenas = texto.split('.') # quebrei as variaveis onde se encontra o '.';
nElementos = len(centenas)

print(nElementos)

for valor in centenas:
  print(valor)

print('##############')

for indice in range(nElementos):
  print(indice,':',centenas[indice])
```
